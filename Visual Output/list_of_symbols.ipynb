```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set Times New Roman globally
rcParams['font.family'] = 'Times New Roman'

# Define full symbol list
symbols_data = [
    ("$i$", "Product index $(i = 1, \\dots, N)$"),
    ("$t$", "Period index $(t = 1, \\dots, T)$"),
    ("$z$", "Forecast horizon"),
    ("$\\hat{D}_{i,t}$", "Forecast of product $i$ for period $t$"),
    ("$D_{i,t}$", "Effective demand of product $i$ for period $t$"),
    ("$S_{i,t}$", "Order-up-to level of product $i$ for period $t$"),
    ("$ss_{i,t}$", "Safety stock of product $i$ for period $t$"),
    ("$I_{i,t}$", "Inventory position of product $i$ for period $t$"),
    ("$Q_{i,t}$", "Quantity ordered for product $i$ at period $t$"),
    ("$LS_{i,t}$", "Lost sales for product $i$ in period $t$"),
    ("$NO_{i,t}$", "Number of orders placed for product $i$ in period $t$ (binary)"),
    ("$SL_i$", "Realized service level for product $i$"),
    ("$TSL_i$", "Target service level for product $i$"),
    ("$Z$", "Z-score from the standard normal distribution"),
    ("$R_i$", "Review period (days) for product $i$"),
    ("$L_i$", "Lead time (days) for product $i$"),
    ("$\\sigma_d$", "Standard deviation of demand during the risk period $(R+L)$"),
    ("$p_i$", "Unit price of product $i$"),
    ("$h$", "Annual holding cost rate (percentage of unit price)"),
    ("$b$", "Lost sales penalty cost rate (percentage of unit price)"),
    ("$k$", "Fixed cost per replenishment order"),
    ("$ADI_i$", "Average inter-demand interval for product $i$"),
    ("$CV_i^2$", "Squared coefficient of variation of product $i$'s demand"),
    ("$RMSE$", "Root Mean Squared Error of forecast"),
    ("$NRMSE$", "Normalized Root Mean Squared Error (relative to mean demand)"),
    ("$RMSSE$", "Root Mean Squared Scaled Error (relative to naive forecast)")
]

# Create DataFrame
df = pd.DataFrame(symbols_data, columns=["Symbol", "Description"])

# Plot
fig, ax = plt.subplots(figsize=(10, 14))
ax.axis("off")

# Create table
table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc="left",
    colLoc="left",
    loc="center",
    edges="horizontal"
)

# Style
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.8)

# Set column widths
table.auto_set_column_width([0, 1])
for i in range(len(df)):
    table[(i + 1, 0)].set_width(0.25)
    table[(i + 1, 1)].set_width(0.75)

# Set header font size
table[0, 0].set_fontsize(14)
table[0, 1].set_fontsize(14)

# Save output
plt.savefig("symbol_table_complete_times_new_roman.png", dpi=300, bbox_inches="tight")
plt.show()
```
