```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set Times New Roman globally
rcParams['font.family'] = 'Times New Roman'

# Define symbols and descriptions
symbols_data = [
    ("$i$", "Product index $(i = 1, \\dots, N)$"),
    ("$t$", "Period index $(t = 1, \\dots, T)$"),
    ("$z$", "Forecast horizon"),
    ("$\\hat{D}_{i,t}$", "Forecast of product $i$ for period $t$"),
    ("$D_{i,t}$", "Effective demand of product $i$ for period $t$"),
    ("$S_{i,t}$", "Order-up-to level of product $i$ for period $t$"),
    ("$ss_{i,t}$", "Safety stock of product $i$ for period $t$"),
    ("$I_{i,t}$", "Inventory position of product $i$ for period $t$"),
    ("$Q_{i,t}$", "Number of items $i$ ordered in period $t$"),
    ("$LS_{i,t}$", "Lost sales of product $i$ for period $t$"),
    ("$NO_{i,t}$", "Number of orders placed for product $i$ in period $t$ (1 or 0)"),
    ("$SL_i$", "Realized service level of product $i$"),
    ("$Z$", "Z-score from the normal distribution"),
    ("$R_i$", "Review period (in days) of product $i$"),
    ("$L_i$", "Lead time of product $i$"),
    ("$TSL_i$", "Target service level of product $i$"),
    ("$p_i$", "Unit value of product $i$"),
    ("$h$", "Annual holding cost, as % of $p_i$"),
    ("$k$", "Cost to place an order"),
    ("$b$", "Unit shortage cost, as % of $p_i$"),
    ("$ADI_i$", "Average inter-demand interval of product $i$"),
    ("$CV_i^2$", "Squared coefficient of variation of demand for product $i$")
]

# Convert to DataFrame
df = pd.DataFrame(symbols_data, columns=["Symbol", "Description"])

# Plot
fig, ax = plt.subplots(figsize=(10, 11))
ax.axis("off")

# Create table
table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc="left",
    colLoc="left",
    loc="center",
    edges="horizontal"
)

# Style
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.8)

# Set column widths
table.auto_set_column_width([0, 1])
for i in range(len(df)):
    table[(i + 1, 0)].set_width(0.2)
    table[(i + 1, 1)].set_width(0.8)
# Set header font size
table[0, 0].set_fontsize(14)
table[0, 1].set_fontsize(14)


# Save output
plt.savefig("symbol_table_times_new_roman.png", dpi=300, bbox_inches="tight")
plt.show()
```
